```python
import matplotlib.pyplot as plt

# Performance data for each program
data = {
    'original_transformer': {
        'L1-dcache-loads': 31169172082,
        'L1-dcache-load-misses': 1426134084,
        'L1-dcache-stores': 2926647907,
        'L1-icache-load-misses': 1898203,
        'LLC-loads': 1092706947,
        'LLC-load-misses': 20202532,
        'dTLB-loads': 31286861764,
        'dTLB-load-misses': 269462139,
        'cycles': 40383249067
    },
    'matrix_transformer': {
        'L1-dcache-loads': 38508989674,
        'L1-dcache-load-misses': 14372623,
        'L1-dcache-stores': 3086197998,
        'L1-icache-load-misses': 1021547,
        'LLC-loads': 285751,
        'LLC-load-misses': 113563,
        'dTLB-loads': 38439948119,
        'dTLB-load-misses': 154762,
        'cycles': 28173139729
    }
}

# Labels and values for the bar graph
labels = list(data['original_transformer'].keys())
original_values = [data['original_transformer'][key] for key in labels]
matrix_values = [data['matrix_transformer'][key] for key in labels]

# Set up the bar width
bar_width = 0.35

# Set up the index for labels
index = range(len(labels))

# Creating the bar graph
plt.figure(figsize=(10, 6))

plt.bar(index, original_values, bar_width, label='Original Transformer')
plt.bar([i + bar_width for i in index], matrix_values, bar_width, label='Matrix Transformer')

# Adding labels and title
plt.xlabel('Performance Metrics')
plt.ylabel('Counts')
plt.title('Performance Comparison Between Original and Matrix Transformer')
plt.xticks([i + bar_width / 2 for i in index], labels, rotation=45)
plt.legend()

# Display the bar graph
plt.tight_layout()
plt.show()
```

```python

# Creating the bar graph with a logarithmic scale for the Y-axis
plt.figure(figsize=(10, 6))

plt.bar(index, original_values, bar_width, label='Original Transformer')
plt.bar([i + bar_width for i in index], matrix_values, bar_width, label='Matrix Transformer')

# Adding labels and title
plt.xlabel('Performance Metrics')
plt.ylabel('Counts (log scale)')
plt.title('Performance Comparison Between Original and Matrix Transformer (Log Scale)')
plt.xticks([i + bar_width / 2 for i in index], labels, rotation=45)
plt.yscale('log')
plt.legend()

# Display the bar graph
plt.tight_layout()
plt.savefig('Performance.png')
plt.show()
```
